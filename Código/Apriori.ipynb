```python
import pandas as pd
from apyori import apriori
```

```python
#encoding = "cp1252", encoding='utf-8', encoding='latin1'etc
base = pd.read_csv("/content/sample_data/Mercado.csv", ",", encoding="ISO-8859-1")
base
```

```python
base.shape
```

```python
transacoes = []
for i in range(len(base)):
  #print(i)
  #print(base_mercado1.values[i, 0])
  transacoes.append([str(base.values[i, j]) for j in range(base.shape[1])])
```

```python
transacoes
```

```python
regras = apriori(transacoes, min_support = 0.6, min_confidence = 0.7)
saida = list(regras) 
print(len(saida))
print(saida)
```
